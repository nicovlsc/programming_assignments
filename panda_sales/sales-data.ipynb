## Medium
### 1. What are the three most popular categories, by total sales?

```python
iv.groupby('category')['price'].sum().sort_values(ascending=False)
```

### 2. What are the total sales attributed to customers over the age of 45?

```python
ages = pd.merge(iv , ct, how = 'outer', left_on = 'customer_id', right_on ='id')

ages_45 = ages.loc[ages['age'] > 45]
ages_45.groupby('age')['price'].sum().sum()
```

### 3. How is the data distributed across different invoice dates? Are there any seasonal trends or patterns?

```python
import matplotlib.pyplot as plt
import numpy as np

ages['invoice_date'] = pd.to_datetime(ages['invoice_date'], dayfirst=True)
ages['month-year'] = ages['invoice_date'].dt.to_period('m')
age_dates = ages.groupby('month-year')['quantity'].sum()
age_dates

age_dates.plot()
plt.xlabel('Date')
plt.ylabel('Sales (Total Quantity)')

#during the pandemic sales would drop during the month of february
#data was recorded in the middle of march so the it looks like a massive decrease in sales but the march data is incomplete
```

## Hard
### 1. Create a pivot table showing the breakdown of sales across these dimensions, in this order: category, decade age range (e.g., 10-19, 20-29, and so forth).

```python
age_range = []
for age_data in ct['age']:
        if age_data >= 10 and age_data < 20:
            age_range.append('10-19')
        elif age_data >= 20 and age_data < 30:
            age_range.append('20-29')
        elif age_data >= 30 and age_data < 40:
            age_range.append('30-39')
        elif age_data >= 40 and age_data < 50:
            age_range.append('40-49')
        elif age_data >= 50 and age_data < 60:
            age_range.append('50-59')
        elif age_data >= 60 and age_data < 70:
            age_range.append('60-69')
        elif age_data >= 70 and age_data < 80:
            age_range.append('70-79')
        elif age_data >= 80 and age_data < 90:
            age_range.append('80-89')
        elif age_data >= 90 and age_data < 100:
            age_range.append('90-99')
        else:
            age_range.append('No age data')

ct['age range'] = age_range

age_sales = iv.merge(ct, how = 'inner', left_on = 'customer_id', right_on = 'id')
pd.pivot_table(age_sales, values = 'price', index = 'category', columns = ['age range'], aggfunc = np.sum)
```
