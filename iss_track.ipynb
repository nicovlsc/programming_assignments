```python
import matplotlib.pyplot as plt
import requests
```

```python
import numpy as np
import time

longitude_values = []
latitude_values = []
timestamp_values = []

h=0
while h < 100:
    r = requests.get("http://api.open-notify.org/iss-now.json")
    longitude_values.append(r.json()['iss_position']['longitude'])
    latitude_values.append(r.json()['iss_position']['latitude'])
    timestamp_values.append(r.json()['timestamp'])
    h += 1
    time.sleep(10)
print(longitude_values)
print(latitude_values)
print(timestamp_values)
```

```python
longitude_float = [float(h) for h in longitude_values]
latitude_float = [float(h) for h in latitude_values]

x = longitude_float
y = latitude_float

plt.title("ISS Coordinates")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.scatter(x, y)

plt.xticks(np.arange(min(longitude_float), max(longitude_float), 10))
plt.yticks(np.arange(min(latitude_float), max(latitude_float), 4))

plt.show()
```
